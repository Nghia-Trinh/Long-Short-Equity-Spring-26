Since the `stock_candles` endpoint returned a 403 error, but `quote` and `company_basic_financials` worked, it indicates that the provided API key has a free-tier limitation that prevents accessing historical candle data and options flow metrics. I will use the `company_basic_financials` endpoint to retrieve a set of alternative price and performance metrics (such as 52-week highs/lows and price returns) to proceed with the strategy, while consolidating these into a structured DataFrame.



```python
import pandas as pd

symbols = ['AAPL', 'MSFT', 'GOOGL']
metrics_data = []

for symbol in symbols:
    try:
        # Fetching basic financials as an alternative since candles are restricted
        res = finnhub_client.company_basic_financials(symbol, 'all')
        if 'metric' in res:
            m = res['metric']
            # Extracting relevant alpha/risk metrics available in the free tier
            metrics_data.append({
                'Symbol': symbol,
                'Price': quote.get('c') if symbol == 'GOOGL' else None, # Using last quote for current price
                '10DayAvgVol': m.get('10DayAverageTradingVolume'),
                '52WeekHigh': m.get('52WeekHigh'),
                '52WeekLow': m.get('52WeekLow'),
                '13WeekPriceReturn': m.get('13WeekPriceReturnDaily'),
                '26WeekPriceReturn': m.get('26WeekPriceReturnDaily'),
                '52WeekPriceReturn': m.get('52WeekPriceReturnDaily'),
                'Beta': m.get('beta')
            })
            print(f"Retrieved metrics for {symbol}")
    except Exception as e:
        print(f"Failed to fetch metrics for {symbol}: {e}")

# Create a consolidated DataFrame
df_metrics = pd.DataFrame(metrics_data)
print('\nConsolidated Metrics Dataset:')
print(df_metrics)
```

## Generate Long-Short Alpha Signals

### Subtask:
Compute the trading signals from the fetched financial data to form an Alpha Matrix.


**Reasoning**:
I will calculate the Momentum_Score, Volatility_Adjusted_Return, and Price_Relative_to_High signals as specified, then aggregate them into an Alpha_Score while handling missing values to form the final Alpha Matrix.



```python
import pandas as pd

# 1. Calculate Momentum_Score
df_metrics['Momentum_Score'] = df_metrics[['13WeekPriceReturn', '26WeekPriceReturn', '52WeekPriceReturn']].mean(axis=1)

# 2. Create Volatility_Adjusted_Return
df_metrics['Volatility_Adjusted_Return'] = df_metrics['13WeekPriceReturn'] / df_metrics['Beta']

# 3. Calculate Price_Relative_to_High
# We use '52WeekHigh' if 'Price' is NaN as per instructions to determine distance from resistance
# Range = High - Low. Score = (Price - Low) / (High - Low)
effective_price = df_metrics['Price'].fillna(df_metrics['52WeekHigh'])
df_metrics['Price_Relative_to_High'] = (effective_price - df_metrics['52WeekLow']) / (df_metrics['52WeekHigh'] - df_metrics['52WeekLow'])

# 4. Combine into Alpha_Score
signal_cols = ['Momentum_Score', 'Volatility_Adjusted_Return', 'Price_Relative_to_High']
# Fill missing values within signal columns with their respective column means
for col in signal_cols:
    df_metrics[col] = df_metrics[col].fillna(df_metrics[col].mean())

df_metrics['Alpha_Score'] = df_metrics[signal_cols].mean(axis=1)

# 5. Print resulting DataFrame
print("--- Alpha Matrix ---")
columns_to_show = ['Symbol', 'Momentum_Score', 'Volatility_Adjusted_Return', 'Price_Relative_to_High', 'Alpha_Score']
print(df_metrics[columns_to_show])
```
